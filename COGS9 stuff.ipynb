```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
from scipy import stats

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = 8, 6
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Tahoma']
```

```python
samples = 10000

def plot_distributions(dist, bins, xlims):
    sns.distplot(dist, bins=bins, kde=False)
    plt.xlim(xlims)
    plt.show()

possible_events = 2
bins = 50
rand_bernoulli = stats.bernoulli.rvs(1/possible_events, size=samples)
plot_distributions(rand_bernoulli, bins, (-0.02, 1.02))

lo, hi = 1, 101
bins = (np.size(np.arange(lo, hi))*2)-1
rand_uniform = stats.randint.rvs(lo, hi, size=samples)
plot_distributions(rand_uniform, bins, ((lo+1)-0.1, hi+0.1))

lo, hi = 1, 101
bins = (((np.size(np.arange(lo, hi))*2)-1)*2)-1
rand_binomial = [np.sum(stats.randint.rvs(lo, hi, size=2)) for i in np.arange(samples)]
plot_distributions(rand_binomial, bins, ((lo+1)-0.1, ((hi-1)*2)+0.1))
```
